```python
import matplotlib.pyplot as plt #для создания двумерных графиков
from mpl_toolkits.mplot3d import Axes3D #для создания трехмерных графиков
```

**Задание №1.** Функция "first_system" возвращает следующую точку двумерного отображения. Работа функции демонстируется выводом 5ти значений, в качестве примера были взяты параметры a0=0.5 и b0=0.6

```python
def first_system(x0, y0, a0, b0):
  x1 = 1.0 - (a0*(x0**2.0)) + y0
  y1 = b0 * x0
  return x1, y1

x0, y0 = 0.01, 0.01
a0, b0 = 0.5,  0.6
for i in range(5):
  if i==0:
    print("x0 y0 = ", x0, y0)
  else:
    x0, y0 = first_system(x0, y0, a0, b0);
    print(f"x{i} y{i} = {x0} {y0}")
```

**Задания №2 и №3.** Начальная точка - (0.01, 0.01). Для каждой пары параметров создаётся отдельный график. Так же создаётся график, объединяющий предыдущие три в одной области. Выводятся первые 15 точек для каждой траектории.

```python
parameters = [(0.8, 0.3), (1.25, 0.04), (1.42, 0.26)]
original_x, original_y = 0.01, 0.01 #начальная точка

colors = ['r', 'g', 'b'] #инициализация цветов
#списки для общего графика
all_x_coords=[]
all_y_coords=[]
k=0

for a, b in parameters:
  plt.figure(figsize=(8,8))
  x, y = original_x, original_y
  #списки для отдельных графиков
  x_coord = []
  y_coord = []
  for i in range(1500):
    x_coord.append(x)
    y_coord.append(y)
    x, y = first_system(x, y, a, b)

  #вывод первых точек траектории
  print(f"Trajectory a={a}, b={b}")
  for i in range(15):
    print(f"x{i}={x_coord[i]}, y{i}={y_coord[i]}")
  #создание отдельных графиков
  plt.plot(x_coord, y_coord, color=colors[k])
  plt.title(f"a={a}, b={b}")
  plt.xlabel('x')
  plt.ylabel('y')
  plt.show()

  all_x_coords.append(x_coord)
  all_y_coords.append(y_coord)
  k+=1

#создание общего графика
plt.figure(figsize=(8,8))
for i in range(len(parameters)):
  plt.plot(all_x_coords[i], all_y_coords[i], color=colors[i])
plt.title("Сombined trajectories")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Задание №4.** Функция "second_system" возвращает следующую точку трехмерного отображения. Создаётся общий график и отдельные для каждого набора параметров. Выводятся первые 15 точек каждой траектории.

```python
def second_system(x, y, z, A, B, C):
  x1 = y
  y1 = z
  z1 = B * x + C * y + A * z - y ** 2.0
  return x1, y1, z1

parametersABC = [(0.61, -0.5, 0.13), (-0.12, -0.5, -0.89), (-0.42, -0.5, -1.38)]
original_z = 0.01

#списки для общего графика
all_x_coords_second_system=[]
all_y_coords_second_system=[]
all_z_coords_second_system=[]

r=0
for A, B, C in parametersABC:
  field = plt.figure(figsize=(8,8))
  x, y, z = original_x, original_y, original_z
  #списки для отдельных графиков
  x_coord = []
  y_coord = []
  z_coord = []
  for i in range(1500):
    x_coord.append(x)
    y_coord.append(y)
    z_coord.append(z)
    x, y, z = second_system(x, y, z, A, B, C)

  #вывод первых точек траектории
  print(f"Trajectory A={A}, B={B}, C={C}")
  for i in range(15):
    print(f"x{i}={x_coord[i]}, y{i}={y_coord[i]}, z{i}={z_coord[i]}")
  #создание отдельных графиков
  three_dimension = field.add_subplot(111, projection = '3d')#добавление оси для трехмерного отображения
  three_dimension.plot(x_coord, y_coord, z_coord, color=colors[r])
  three_dimension.set_title(f"A={A}, B={B}, C={C}")
  three_dimension.set_xlabel('x')
  three_dimension.set_ylabel('y')
  three_dimension.set_zlabel('z', labelpad=0) #отступ = 0, чтобы уместилась буква z на графике
  plt.show()

  all_x_coords_second_system.append(x_coord)
  all_y_coords_second_system.append(y_coord)
  all_z_coords_second_system.append(z_coord)
  r+=1

#создание общего графика
field = plt.figure(figsize=(8,8))
three_dimension = field.add_subplot(111, projection='3d')
for i in range(len(parameters)):
  three_dimension.plot(all_x_coords_second_system[i], all_y_coords_second_system[i], all_z_coords_second_system[i], color=colors[i])
three_dimension.set_title("Сombined trajectories")
three_dimension.set_xlabel("x")
three_dimension.set_ylabel("y")
three_dimension.set_zlabel("z", labelpad=0)
plt.show()
```

**Задание №5.** Функция "logistic_display" реализует логистическое отображение. Создается 800 значений параметра r в промежутке от 2.4 до 4.0. Изначально значение x=0.01. Для каждого значения параметра r выполняется 10.000 итераций, последняя 1000 значений x записываются в массив, на основе которого визуализируется дерево.  


```python
def logistic_display(r, x):
  result = r * x * (1 - x)
  return result

#заполнение списка с параметрами
r_values=[]
r_start = 2.4
r_end = 4.0
r_step = (r_end - r_start)/800
r=r_start
while r < r_end:
  r_values.append(r);
  r += r_step

#заполнение списка со значениями
x_values = []
for r in r_values:
  x = original_x
  for i in range (10000):
    x = logistic_display(r, x)
    if i>=9000:
      x_values.append(x)

#список для сопоставления параметра и значения для создания дерева
repeated_r_values=[]
for r in r_values:
  repeated_r_values.extend([r]*1000)

#создание дерева
plt.figure(figsize=(8, 8))
plt.plot(repeated_r_values, x_values, 'g.', alpha=0.1)
plt.title('Bifurcation tree')
plt.xlabel('r')
plt.ylabel('x')
plt.xlim(2.4, 4.0)
plt.ylim(0, 1)
plt.show()
```
